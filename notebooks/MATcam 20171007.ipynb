```python
import os
import sys
import glob

import numpy as np
import matplotlib
matplotlib.use('nbagg')
#from matplotlib import style
#style.use('ggplot')
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2

import astropy.units as u
from astropy.io import fits
from astropy import visualization

import ccdproc
```

```python
def wfs_norm(data, interval=visualization.MinMaxInterval(), stretch=visualization.LinearStretch()):
    """
    Define default image normalization to use for WFS images
    """
    norm = visualization.mpl_normalize.ImageNormalize(
        data,
        interval=interval,
        stretch=stretch
    )
    return norm

def implot(data):
    fig, ax = plt.subplots()
    ax.imshow(data, cmap='Greys', origin='lower', norm=wfs_norm(data), interpolation='None')
    return fig, ax
```

```python
ic = ccdproc.ImageFileCollection("../pointing_data/20171007")
```

```python
images = list(ic.ccds(ccd_kwargs={"unit": u.adu}))
```

```python
fig, ax = implot(images[20])
fig.show()
```
